# **DATA AUGMENTATION**

# **Data augmentation is the process of artificially generating new data from existing data, primarily to train new machine learning (ML) models.**

```python
# impoerting dependencies

from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

```python
# loading image from which we will form more images
img=image.load_img('/content/dog.webp',target_size=(200,200))
```

```python
#image preview
plt.imshow(img)
```

```python
# type of operations we want to perform on image
datagen=ImageDataGenerator(
    rotation_range=40, # rotation angle
    width_shift_range=0.2, # shift towards left-right
    height_shift_range=0.2, # shift up-down
    shear_range=0.2,
    zoom_range=0.2, # zoom in and zoom out
    horizontal_flip=True # mirror reflection
)
```

```python
img=image.img_to_array(img) # image to array conversion
```

```python
img_batch=img.reshape(1,200,200,3); # conversion to batch of images
```

```python
i=0;
for output in datagen.flow(img_batch,batch_size=1,save_to_dir='augmentati'): # directory name in which we want to save new images
  i+=1;
  if i>5: # want 5 images
    break;
```
